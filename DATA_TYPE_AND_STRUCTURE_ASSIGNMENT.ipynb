Q4-: Describe how dictionaries store data.
>>Ans-: Dictionaries store data as key-value pairs, where each unique key maps directly to a specific value. They use a mechanism called a hash table to achieve extremely fast data retrieval, regardless of how large the dictionary grows.
               :-Mechanism-:
(1)Hashing Function: When you insert a key, the system passes it through a mathematical function that converts it into a unique integer hash code.(2)Index Mapping: This hash code is converted into an index number that points directly to a specific bucket or slot in an underlying array.
(3)Value Storage: The key, hash code, and value are stored together in that mapped array slot.
(4)Collision Resolution: If two different keys accidentally generate the same index (a collision), the system resolves it by chaining items or finding the next open slot.               

Q5-:Why might you use a set instead of a list in Python.
>>Ans-: You would use a set instead of a list when you need fast item lookups or require a collection of completely unique elements.
(1)Instant Search Speed (O(1) vs O(N))
>>List: Python must scan a list item by item from the beginning. If the list has 10 million items, finding the last item takes 10 million steps.
>>Set: Sets use a hash table. Python calculates the item's location instantly, taking the exact same amount of time whether the set has 5 items or 5 million items.
(2) Automatic Duplication Removal
>>List: Allows duplicate values. You have to write custom logic to check for and remove duplicates.
>>Set: Strictly enforces uniqueness. If you add 5 to a set that already contains 5, it silently ignores the duplicate.
(3) Built-in Mathematical OperationsSets provide highly optimized methods to compare collections instantly, which require complex loops if done with lists:
>>Intersection (&): Find items present in both collections.
>>Union (|): Combine both collections while stripping duplicates.
>>Difference (-): Find items present in one collection but not the other.

Q8-: What is a hash table, and how does it relate to dictionaries in Python?
>>Ans-: A hash table is a foundational data structure that maps unique keys to values using a mathematical formula called a hashing function. In Python, a dictionary is simply a direct implementation of a hash table.
 >> How does it relate to dictionaries in Python-:
 (1)>>Time Complexity: Because the hash function tells Python exactly where to look in memory, retrieving my_dict["Alice"] happens instantly. Python does not need to scan through the dictionary line by line.
 (2)>>The Immutability Rule: Dictionary keys must be immutable (like strings, numbers, or tuples). If a key's value could change, its hash code would change, and Python would look in the wrong memory slot, losing your data.
 (3)>>Hash Collisions: Occasionally, two completely different keys produce the exact same memory index. Python dictionaries handle this using a technique called open addressing (specifically, pseudo-random probing) to find the next available empty slot in the table.
 (4)>>Compact Layout: Modern Python dictionaries (since Python 3.6) use a two-array system. One small array stores the hash indices, and a second tight array stores the actual data sequentially. This makes Python dictionaries incredibly memory efficient and allows them to remember the order in which items were added.
